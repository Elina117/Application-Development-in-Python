```python
# Шаг 1: Отфильтруем данные для ресторанов Taco Bell
taco_bell_df = df[df['name'] == 'Taco Bell']


# Шаг 2: Подсчет количества ресторанов Taco Bell в каждом городе
city_counts = taco_bell_df['city'].value_counts()


# Шаг 3: Получение топ-5 городов с наибольшим количеством ресторанов Taco Bell
top_5_cities = city_counts.head(5).index

# Шаг 4: Фильтрация данных только по топ-5 городам
# Применяем isin() только к столбцу 'city'
filtered_df = taco_bell_df[taco_bell_df['city'].isin(top_5_cities)]

# Шаг 5: Подсчет количества ресторанов Taco Bell в этих городах и сохранение в Series
result = filtered_df['city'].value_counts()
print(result)
```

```python
#преобразовываем данные о дате в дэйт тайм
df['dateAdded'] = pd.to_datetime(df['dateAdded'])

#группируем данные по месяцу и извлекаем уникальный айди ресторана
result = df.groupby(df['dateAdded'].dt.month)['id'].nunique()
result = result.to_frame()
# #устанавливаем индексы для формата датафрэйм
print(result)
```

```python
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df['update_delta'] = (df['dateUpdated'] - df['dateAdded']).dt.days

zep_mean = df[df['city'] == 'Zephyrhills']['update_delta'].mean()

result = df.groupby('city').agg(
    {
        'update_delta': 'mean', 
        'latitude':'max'
    }).reset_index()
print(result)
```
