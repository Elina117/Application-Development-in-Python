```python
#  1:     Taco Bell
taco_bell_df = df[df['name'] == 'Taco Bell']


#  2:    Taco Bell   
city_counts = taco_bell_df['city'].value_counts()


#  3:  -5      Taco Bell
top_5_cities = city_counts.head(5).index

#  4:     -5 
#  isin()    'city'
filtered_df = taco_bell_df[taco_bell_df['city'].isin(top_5_cities)]

#  5:    Taco Bell       Series
result = filtered_df['city'].value_counts()
print(result)
```

```python
df['dateUpdated'] = pd.to_datetime(df['dateUpdated'])
df['dateAdded'] = pd.to_datetime(df['dateAdded'])
df['update_delta'] = (df['dateUpdated'] - df['dateAdded']).dt.days

zep_mean = df[df['city'] == 'Zephyrhills']['update_delta'].mean()

result = df.groupby('city').agg(
    {
        'update_delta': 'mean', 
        'latitude':'max'
    }).reset_index()
print(result)
```
