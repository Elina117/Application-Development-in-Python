```python
uniq_apps = df['App'].nunique()
print(uniq_apps)
```

```python
rating_missing_cells = df[df['Rating'].isna()]
rating_missing = len(rating_missing_cells)
print(rating_missing)
```

```python
row_concat = pd.concat([
    df.loc[0:2],
    df.loc[5:7],
    df.loc[15:18]])

columns_concat = row_concat[['App', 'Size', 'Genres', 'Current Ver']]
columns_concat.to_csv('concat_cav_file.csv', index=True)
print(columns_concat)
```

```python
df_1 = df.drop_duplicates(subset=['App'])
unique_playstore = df_1.reset_index(drop=True)
print(unique_playstore)
df.columns
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
# Удаляем дубликаты по колонке 'app'
df_1 = df.drop_duplicates(subset=['app'])

total_apps = len(df_1)
print(total_apps)
free_apps = (df_1['price'] == '0').sum()
result = (free_apps / total_apps).round(2)

print(result)
```

```python
education_playstore = df[
    (df['category'] == 'EDUCATION') & (df['reviews'] > 1000)
].reset_index(drop=True)
print(education_playstore)
```

```python
df['price'] = df['price'].str.replace('$', '', regex = True).astype('float')
df.columns
```

```python
df = df.drop_duplicates(subset=['app'])

df['price'] = df['price'].astype(float)
df['rating'] = df['rating'].astype(float)
df['reviews'] = df['reviews'].astype(float)


# Создаем сводную таблицу
pivot_table = pd.pivot_table(
    df,
    values=['price', 'rating', 'reviews'],
    index=['category', 'type'],
    aggfunc={'price': 'mean', 'rating': 'mean', 'reviews': 'mean'}
)
# Переименовываем колонки
pivot_table.columns = ['mean_price', 'mean_rating', 'mean_reviews']

# Округляем значения
pivot_table['mean_price'] = pivot_table['mean_price'].round(2)
pivot_table['mean_rating'] = pivot_table['mean_rating'].round(1)
pivot_table['mean_reviews'] = pivot_table['mean_reviews'].round(2)

pivot_table.to_csv('pivot_table_apps.csv', sep = ',')
df.columns
```
